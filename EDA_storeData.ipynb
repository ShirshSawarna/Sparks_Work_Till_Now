```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
fig = px.histogram(df, x = 'Profit')
fig.show()
```

### Distribution of Columns (Numerical and Categorical)

```python
df_categorical = df.select_dtypes(include = 'object').columns

for i in df_categorical:
    fig = px.histogram(df, x = i,  color_discrete_sequence=['indianred'])
    fig.show()
```

```python
df_numerical = df.select_dtypes(exclude = 'object').columns

for i in df_numerical:
    fig = px.histogram(df, x = i)
    fig.show()
```

###Boxplot for Outlier Check

```python
for i in df_numerical:
    fig = px.box(df, y=i)
    fig.show()
```

### Variance of Profit with categorical columns

```python
for i in df_categorical:
    fig = px.box(df, y = 'Profit', color = i)
    fig.show()
```

### Pairplots

```python
fig = px.scatter_matrix(df[df_numerical], color_discrete_sequence=['dodgerblue'])
fig.show()
```

### Correlation Heatmap

```python
fig = px.imshow(df.corr())
fig.show()
```
